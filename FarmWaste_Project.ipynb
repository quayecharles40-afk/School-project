### Waste Management system

```python
from datetime import date
from abc import ABC, abstractmethod



class FarmWaste(ABC):    #Parent class
    def __init__(self, quantity):
        self.__quantity = quantity        #Encapsulation
        self.__date_recorded = date.today()

    def get_quantity(self):               # Using the getter method
        return self.__quantity
    
    def get_date_recorded(self):
        return self.__date_recorded
    
    def set_quantity(self, new_quantity):   # Using the setter method
        if new_quantity <= 0:
            raise ValueError("Quantity must be positive")
        self.__quantity = new_quantity

    @abstractmethod
    def determine_disposal(self):      
        pass

    @abstractmethod
    def get_waste_type(self):
        pass
    
    @classmethod
    def create_waste(cls, waste_type, quantity):
        if waste_type == "1":
            return OrganicWaste(quantity)
        elif waste_type == "2":
            return ChemicalWaste(quantity)
        elif waste_type == "3":
            return InfectedCropWaste(quantity)
        elif waste_type == "4":
            return PlasticContainers(quantity)
        else:
            raise ValueError("Invalid waste type")


class OrganicWaste(FarmWaste):
    def determine_disposal(self):
        return "Compost it to produce fertilizer"
    
    def get_waste_type(self):    #Polymorphism
        return "Organic Waste"


class ChemicalWaste(FarmWaste):
    def determine_disposal(self):
        return "Store safely and hand over to hazardous waste management"
    
    def get_waste_type(self):
        return "Chemical Waste"


class InfectedCropWaste(FarmWaste):
    def determine_disposal(self):
        return "Burn infected crops to prevent disease spread"
    
    def get_waste_type(self):
        return "Infected Crop Waste"


class PlasticContainers(FarmWaste):
    def determine_disposal(self):
        return "Clean and recycle or dispose properly"
    
    def get_waste_type(self):
        return "Plastic Containers"


# ==================================================
#   MENU
# ==================================================

all_waste = []   

print("Welcome to Farm Waste Helper!")

while True:
    print("\nWhat do you want to do?")
    print("1 → Add organic waste")
    print("2 → Add chemical waste")
    print("3 → Add infected crop waste")
    print("4 → Add plastic containers")
    print("5 → See all waste I saved")
    print("6 → Finish and exit")

    choice = input("Type number 1–6: ")

    if choice == "6":
        print("\nGoodbye 👋")
        break

    if choice == "5":
        # Show everything we saved
        if len(all_waste) == 0:
            print("\nNothing saved yet...")
        else:
            print("\nHere is everything you saved:")
            number = 1
            for item in all_waste:
                print(number, item.get_waste_type())
                print("   Quantity:", item.get_quantity())
                print("   Date:    ", item.get_date_recorded())
                print("   What to do:", item.determine_disposal())
                number = number + 1 #or number += 1
        continue

    if choice not in ["1","2","3","4"]:
        print("Please enter only 1, 2, 3, 4, 5 or 6")
        continue

    # Ask for quantity and keep asking until number is okay
    while True:
        try:
            quantity = float(input("Please enter the quanitity? (number please): "))
               # try to turn it into number instead of a string

            if quantity <= 0:
                print("Please write a number bigger than 0")
                continue   # ask again

            
            break

        except:
            print("That's not a number... please try again (example: 5 or 12.5)")

    # Now we create the waste and save it
    new_waste = FarmWaste.create_waste(choice, quantity)
    all_waste.append(new_waste)

    print("\nSaved!")
    print("Type:", new_waste.get_waste_type())
    print(f"{quantity} Kg")
    print("What to do:", new_waste.determine_disposal())
    print("Date:", new_waste.get_date_recorded())

print("\nThank you for using the program!")
```
